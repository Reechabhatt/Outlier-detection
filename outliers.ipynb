```python
import pandas as pd
data = pd.read_csv("domain3.csv")
```

```python
#detecting and removing outlier
import matplotlib.pyplot as plt
plt.figure()
plt.boxplot(data["words"],0, 'gD')
plt.show()
```

```python
q3 = data["words"].quantile(0.75)
q3
q4 = data["words"].quantile(0.25)
q4
iqr=q3-q4
iqr
```

```python
ul=q3+(1.5*iqr)
ll=q4-(1.5+iqr)
```

```python
ul
```

```python
ll
```

```python
data[((data["words"] < ll) |(data["words"] > ul))]
```

```python
data.drop([189,255,302,1211,1363])
```

```python
 data.to_csv('cleansemen.csv')
```
